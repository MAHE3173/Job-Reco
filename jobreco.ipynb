```python
selected_features = ['jobTitle','jobClassification','state','city','workType','language']
#data that is combined from the selected features for job searching
data2 = data1['jobTitle']+' '+data1['jobClassification']+' '+data1['state']+' '+data1['city']+' '+data1['workType']+ ' '+data1['language']
```

```python
#text cleaning and tokenizing the combined information of job description 
teaser = data2.tolist()
from string import punctuation
from nltk.corpus import stopwords
stop_nltk = stopwords.words("english")
punct_stop = list(punctuation) + stop_nltk
from nltk.tokenize import word_tokenize
desc_tokens=[]
for desc in teaser:
    desc_tokens.append([word.lower() for word in word_tokenize(desc) if word not in punct_stop])
print(desc_tokens[0])
```

```python
#input information for job searching
search = str(input('enter job criteria: '))
search.lower()
```

```python
desc_str=[]
for temp in desc_tokens:
    desc_str.append(' '.join(temp)) #making description strings from tokens
    
desc_str.append(search)    # adding input information to the description strings
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()

feature_vectors = vectorizer.fit_transform(desc_str) #construct TF-IDF Matrix 
similarity = cosine_similarity(feature_vectors)

similarity_score = list(enumerate(similarity[1498]))
sorted_similarity = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
del sorted_similarity[0]
print(sorted_similarity) #the sorted values of similarity of given dataset with our input string
```

```python
#print out 20 jobs suitable for the user
print('jobs suggested for you : \n')

i = 1
templist =[]
for job in sorted_similarity:
  index = job[0]
  title_from_index = data1[data1.index==index].values[0]
  if (i<51):
    templist.append(title_from_index)
    i+=1
    
jobs = pd.DataFrame(templist, columns = data1.columns.tolist())
jobs
```
